```python
import pandas as pd
import ast

# Load the Excel file
df = pd.read_excel(r"C:\Users\VIGNESH\Downloads\bangalore_cars.xlsx")

# Function to flatten the new_car_specs column
def flatten_specs(specs):
    flat_data = {}
    try:
        specs_dict = ast.literal_eval(specs)  # Convert string representation of dict to actual dict
        # Flatten top-level keys
        for item in specs_dict['top']:
            flat_data[item['key']] = item['value']
        # Flatten data section
        for data in specs_dict['data']:
            for detail in data['list']:
                flat_data[detail['key']] = detail['value']
    except (ValueError, SyntaxError):
        print(f"Error parsing specs: {specs}")
    return flat_data

# Apply the flattening function to the new_car_specs column
flattened_data3 = df['new_car_specs'].apply(flatten_specs).apply(pd.Series)

# Display the flattened DataFrame
print(flattened_data3.head())
```

```python
flattened_data3
```

```python
df_3 = pd.DataFrame(flattened_data3)
df_3['City'] = 'Chennai'
df_3=df_3.drop('Ground Clearance Unladen',axis=1)
df_3=df_3.drop('Wheel Size',axis=1)
df_3=df_3.drop('Max Torque',axis=1)
df_3=df_3.drop('Seats',axis=1)
df_3 = df_3.rename(columns={'Mileage': 'Mileage(Kmpl)'})
df_3=df_3.drop('Engine',axis=1)
df_3['Mileage(Kmpl)'] = df_3['Mileage(Kmpl)'].str.lower().str.replace('kmpl', '', regex=False).str.strip()
df_3['Max Power'] = df_3['Max Power'].str.replace(r'b[hH][pP]@\d+[rR][pP][mM]', '', regex=True).str.strip()
df_3 = df_3[['Mileage(Kmpl)', 'Max Power', 'Color', 'City']]
```

```python
df_3
```

```python
df_bangalore = pd.concat([df, df_1, df_3], axis=1)
df_bangalore =df_bangalore.rename(columns={'Max Power': 'Max Power(bhp)'})
df_bangalore
```

```python
def clean_data1(value):
    # Remove commas and strip whitespace
    if isinstance(value, str):
        value = value.replace(',', '').strip()

        # Convert "Crore" to numeric (1 Crore = 10,000,000)
        if 'Crore' in value:
            value = float(value.replace('Crore', '').strip()) * 1e7  # Convert to numeric value
        
        # Convert "Lakh" to numeric (1 Lakh = 100,000)
        elif 'Lakh' in value:
            value = float(value.replace('Lakh', '').strip()) * 1e5  # Convert to numeric value

    # Handle empty strings by converting them to 0
    if value == '':
        value = 0

    # Convert to float
    return float(value)
```

```python
import re

def clean_power(value):
    if isinstance(value, str):
        # Extract numeric part using a regular expression
        match = re.search(r'\d+(\.\d+)?', value)  # Find numbers (including decimals)
        if match:
            value = match.group(0)  # Get the first matched number
        else:
            value = 0  # If no number is found, set to 0

    # Convert to float
    try:
        return float(value)
    except ValueError:
        return None  # Return None if conversion fails
```
